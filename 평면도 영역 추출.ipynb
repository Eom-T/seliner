cv2.RETR_TREE : 모든 contour와 상관관계 추출
cv2.RETR_EXTERNAL : 이미지의 가장 바깥쪽 contour만 추출
cv2.RETR_LIST : 상관관계를 고려하지 않고 contour추출
cv2.RETR_COOMP : 이미지에서 모든 contour 추출 후 2단계 contour 계층구조로 구성, 1단계는 외곽경계, 2단계는 구멍

cv2.CHAIN_APPORX_NONE : contour를 구성하는 모든 점을 저장
cv2.CHAIN_APPORX_SIMPLE : contour의 수평, 수직, 대각선 방향의 점은 모두 버리고 끝 점만 남겨둠
cv2.CHAIN_APPORX_TC89__1 : Teh=chin 연결 근사 알고리즘 적용

```python
import numpy as np
import cv2 
import time
import matplotlib.pyplot as plt


start = time.time()
img = cv2.imread('1_b.jpg')
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thr = cv2.threshold(imgray, 127, 255, 0)

plt.imshow(cv2.cvtColor(thr, cv2.COLOR_GRAY2RGB))
plt.show()

contours, hierarchy = cv2.findContours(thr, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

img = cv2.drawContours(img, contours, -1, (0, 0, 255), 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
t = time.time() - start
print(t)
```
